```python
import pandas
old_dataset = pandas.read_csv("new_dataset_old.csv",index_col=0)
dataset = pandas.read_csv("dataset_with_img.csv",index_col=0)
```

```python
old_dataset["img_id"] = old_dataset.index
```

```python
dataset = dataset if "img_id" not in dataset.columns else dataset.sort_values(by="img_id")

dataset = dataset[dataset["img_id"].apply(lambda x: x != -1)]
```

```python
base = len(dataset)
rate = lambda x: int(base * sum((6,2,2)[0:{"train":0,"value":1,"test":2}[x]+1])/sum((6,2,2)))


train_set = dataset.iloc[:rate("train")]
value_set = dataset.iloc[rate("train") : rate("value")]
test_set = dataset.iloc[rate("value"):]
```

```python
set_list = (train_set,value_set,test_set)
```

```python
total_set = pandas.DataFrame()
for sets in set_list:
    s = sets.append(old_dataset.iloc[sets["img_id"]])
    total_set = s if total_set.empty else total_set.append(s)

total_set.to_csv("test.csv")
```

```python
old_dataset.drop(index=list(dataset["img_id"])).to_csv("july_data.csv")
```
